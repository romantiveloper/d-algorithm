```python
# N = 1   1
# N = 2   2
# N = 3   3
# N = 4   5
# N = 5   8
# N=[1,2,3,4,5....]
# dp = [1,1,1,1,1...]
# dp=[1,2,3,5,8...]
# [1,1,2,3,5,8] 

# 규칙찾기
# N = 3
# 111
# 100
# 001

# N = 4
# 1001
# 0011
# 1100
# 0000
# 1111

# N = 5
# 10011
# 11001
# 11100
# 00111
# 10000
# 00001
# 00100
# 11111

# --> 피보나치 수열

# 풀이1)
N = int(input())
dp = [1] * 1000001
dp[0] = 1
dp[1] = 1
dp[2] = 2

for i in range(3, N + 1):
    dp[i] = (dp[i - 1] + dp[i - 2]) % 15746
print(dp[N])

# 풀이2)
N = int(input()) 
dp = [1] * (N+1)
if len(dp) >= 3 : 
    dp[2] = 2
    
for i in range(3, N + 1):
    dp[i] = dp[i - 1] + dp[i - 2]
print(dp[N])
```
